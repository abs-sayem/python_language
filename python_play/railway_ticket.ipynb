```python
# Import necessary libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Set up the web driver
# Make sure you have the appropriate driver (e.g., chromedriver) installed
driver = webdriver.Chrome()

try:
    # Step 1: Open the website
    driver.get("https://eticket.railway.gov.bd/")
    print("Website opened successfully")

    # Step 2: Log in to the website
    # Locate and interact with login elements
    username_field = driver.find_element(By.ID, "01825507837")  # Update with actual ID
    password_field = driver.find_element(By.ID, "12111994")  # Update with actual ID
    login_button = driver.find_element(By.ID, "login_button_id")  # Update with actual ID

    username_field.send_keys("01825507837")  # Replace with your username
    password_field.send_keys("12111994")  # Replace with your password
    login_button.click()
    print("Logged in successfully")

    # Step 3: Search for tickets
    # Fill in search form with required journey details
    departure_station = driver.find_element(By.ID, "Chattogram")  # Update with actual ID
    arrival_station = driver.find_element(By.ID, "Laksam")  # Update with actual ID
    departure_date = driver.find_element(By.ID, "departure_date_input_id")  # Update with actual ID
    search_button = driver.find_element(By.ID, "search_button_id")  # Update with actual ID

    departure_station.send_keys("Chattogram")  # Example departure
    arrival_station.send_keys("Laksam")  # Example destination
    departure_date.send_keys("2025-03-29")  # Example date
    search_button.click()
    print("Ticket search initiated")

    # Step 4: Select a ticket
    # Update element selectors based on the website's structure
    wait = WebDriverWait(driver, 10)
    ticket_option = wait.until(EC.element_to_be_clickable((By.XPATH, "ticket_option_xpath")))  # Update XPath
    ticket_option.click()
    print("Ticket selected")

    # Step 5: Complete the purchase
    purchase_button = driver.find_element(By.ID, "purchase_button_id")  # Update with actual ID
    purchase_button.click()
    print("Ticket purchase completed")

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    # Step 6: Close the browser
    driver.quit()
    print("Browser closed")
```
